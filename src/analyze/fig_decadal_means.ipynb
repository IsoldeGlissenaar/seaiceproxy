# Decadal means

Plot proxy SIT decadal means for November and April and seasonal growth

```python
import numpy as np
import xarray as xr
import shapefile as shp  # Requires the pyshp package
from shapely.geometry import Point, shape
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib
land_50m = cfeature.NaturalEarthFeature('physical', 'land', '50m')
```

```python
fig = plt.figure(figsize=(15.8,10.0),dpi=400)
gs1 = gridspec.GridSpec(3,3)
gs1.update(wspace=0.00, hspace=0.05)

m=11
direc = '../../data/proxy_SIT/'
sit = xr.open_dataset(direc+'predic_sit_19932020_'+f"{m:02}"+'.nc')

# November 1992-1999
ax = plt.subplot(gs1[0], projection=ccrs.Orthographic(central_longitude=-99, central_latitude=70, globe=None))
ax.coastlines(resolution='50m',linewidth=0.5)
ax.set_extent([-140,-57,62,84],crs=ccrs.PlateCarree())
ax.gridlines(linewidth=0.3, color='k', alpha=0.5, linestyle=':')
im = plt.scatter(sit.lon, sit.lat, c=np.nanmean(sit.sit_mean[0:8,:].values,axis=0), cmap='Spectral_r',vmin=0,vmax=3,s=6, transform=ccrs.PlateCarree())
ax.add_feature(land_50m, facecolor='#eeeeee')
ax.text(0.03,0.05,'(a)',fontsize=12,transform=ax.transAxes)

# November 2000-2009
ax = plt.subplot(gs1[1], projection=ccrs.Orthographic(central_longitude=-99, central_latitude=70, globe=None))
ax.coastlines(resolution='50m',linewidth=0.5)
ax.set_extent([-140,-57,62,84],crs=ccrs.PlateCarree())
ax.gridlines(linewidth=0.3, color='k', alpha=0.5, linestyle=':')
im = plt.scatter(sit.lon, sit.lat, c=np.nanmean(sit.sit_mean[8:18,:].values,axis=0), cmap='Spectral_r',vmin=0,vmax=3,s=6, transform=ccrs.PlateCarree())
ax.add_feature(land_50m, facecolor='#eeeeee')
ax.text(0.03,0.05,'(b)',fontsize=12,transform=ax.transAxes)

# November 2010-2020
ax = plt.subplot(gs1[2], projection=ccrs.Orthographic(central_longitude=-99, central_latitude=70, globe=None))
ax.coastlines(resolution='50m',linewidth=0.5)
ax.set_extent([-140,-57,62,84],crs=ccrs.PlateCarree())
ax.gridlines(linewidth=0.3, color='k', alpha=0.5, linestyle=':')
im = plt.scatter(sit.lon, sit.lat, c=np.nanmean(sit.sit_mean[18:29,:].values,axis=0), cmap='Spectral_r',vmin=0,vmax=3,s=6, transform=ccrs.PlateCarree())
ax.add_feature(land_50m, facecolor='#eeeeee')
cbar = fig.colorbar(im, ax=ax, label='sea ice thickness [m]',extend='max', fraction=0.046, pad=0.03)
ax.text(0.03,0.05,'(c)',fontsize=12,transform=ax.transAxes)


m=4
direc = '../../data/proxy_SIT/'
sit = xr.open_dataset(direc+'predic_sit_19932020_'+f"{m:02}"+'.nc')

# April 1992-1999
ax = plt.subplot(gs1[3], projection=ccrs.Orthographic(central_longitude=-99, central_latitude=70, globe=None))
ax.coastlines(resolution='50m',linewidth=0.5)
ax.set_extent([-140,-57,62,84],crs=ccrs.PlateCarree())
ax.gridlines(linewidth=0.3, color='k', alpha=0.5, linestyle=':')
im = plt.scatter(sit.lon, sit.lat, c=np.nanmean(sit.sit_mean[0:8,:].values,axis=0), cmap='Spectral_r',vmin=0,vmax=3,s=6, transform=ccrs.PlateCarree())
ax.add_feature(land_50m, facecolor='#eeeeee')
ax.text(0.03,0.05,'(d)',fontsize=12,transform=ax.transAxes)
cbar.ax.locator_params(nbins=4)

# April 2000-2009
ax = plt.subplot(gs1[4], projection=ccrs.Orthographic(central_longitude=-99, central_latitude=70, globe=None))
ax.coastlines(resolution='50m',linewidth=0.5)
ax.set_extent([-140,-57,62,84],crs=ccrs.PlateCarree())
ax.gridlines(linewidth=0.3, color='k', alpha=0.5, linestyle=':')
im = plt.scatter(sit.lon, sit.lat, c=np.nanmean(sit.sit_mean[8:18,:].values,axis=0), cmap='Spectral_r',vmin=0,vmax=3,s=6, transform=ccrs.PlateCarree())
ax.add_feature(land_50m, facecolor='#eeeeee')
ax.text(0.03,0.05,'(e)',fontsize=12,transform=ax.transAxes)
cbar.ax.locator_params(nbins=4)

# April 2010-2020
ax = plt.subplot(gs1[5], projection=ccrs.Orthographic(central_longitude=-99, central_latitude=70, globe=None))
ax.coastlines(resolution='50m',linewidth=0.5)
ax.set_extent([-140,-57,62,84],crs=ccrs.PlateCarree())
ax.gridlines(linewidth=0.3, color='k', alpha=0.5, linestyle=':')
im = plt.scatter(sit.lon, sit.lat, c=np.nanmean(sit.sit_mean[18:29,:].values,axis=0), cmap='Spectral_r',vmin=0,vmax=3,s=6, transform=ccrs.PlateCarree())
ax.add_feature(land_50m, facecolor='#eeeeee')
ax.text(0.03,0.05,'(f)',fontsize=12,transform=ax.transAxes)
cbar = fig.colorbar(im, ax=ax, label='sea ice thickness [m]',extend='max', fraction=0.046, pad=0.03)
cbar.ax.locator_params(nbins=4)


#---------------------------------------------------------------------------------------------

months = [11,12,1,2,3,4]
years = np.arange(1992,2021,1)
direc = '../../data/proxy_SIT/'

# Check in which CISIRR Region the grid cells are
idx = []
shp1 = shp.Reader('../../data/misc/v200_CISIRR_Regions_4326_merge.shp') #open the shapefile
all_shapes = shp1.shapes() # get all the polygons
all_records = shp1.records()     

predictions_monthmean = np.zeros((29,6))
predictions_monthmean[:,:] = np.nan
for i in range((6)):
    m = months[i]
    sit = xr.open_dataset(direc+'predic_sit_19932020_'+f"{m:02}"+'.nc')
    
    #Get location
    len_f = sit.dims['n']
    location = []; p=0
    for n in range(len_f):
        pt = (sit.lon.values[n], sit.lat.values[n])  
        for k in range (len(all_shapes)):             
            boundary = all_shapes[k]
            if Point(pt).within(shape(boundary)):
                location.append(all_records[k][3])
                p=1
        if p==0:
            location.append(' ')
        p=0
    sit['location'] = (['n'], np.array(location))
    #Remove outside Canadian Arctic area (to avoid changing shapes)
    idx_ca = np.where(sit.location.values==' ')[0]
    sit.sit_mean.values[:,idx_ca] = np.nan
    
    for j in range(len(sit.year.values)):
        idx = np.where(years==sit.year.values[j])
        predictions_monthmean[idx,i] = np.nanmean(sit.sit_mean.values[j,:][sit.lon.values>-156.6])

cmap = matplotlib.cm.get_cmap('viridis')

predictions_1992 = np.nanmean(predictions_monthmean[0:8,:],axis=0)
predictions_2001 = np.nanmean(predictions_monthmean[8:18,:],axis=0)
predictions_2011 = np.nanmean(predictions_monthmean[18:29,:],axis=0)
predictions_all = np.nanmean(predictions_monthmean[:,:],axis=0)

predictions_1992_std = np.nanstd(predictions_monthmean[0:8,:],axis=0)
predictions_2001_std = np.nanstd(predictions_monthmean[8:18,:],axis=0)
predictions_2011_std = np.nanstd(predictions_monthmean[18:29,:],axis=0)

dec = ['1992-1999', '2000-2009','2010-2020']

ax = plt.subplot(gs1[7])
plt.plot(predictions_1992, color='#FF6501', label=str(dec[0]))
plt.plot(predictions_2001, color='#009901', label=str(dec[1]))
plt.plot(predictions_2011, color='#670099', label=str(dec[2]))
plt.fill_between(np.arange(0,6,1), predictions_1992-predictions_1992_std,
                  predictions_1992+predictions_1992_std, color='#FF6501',alpha=0.15)
plt.fill_between(np.arange(0,6,1), predictions_2001-predictions_2001_std,
                  predictions_2001+predictions_2001_std, color='#009901',alpha=0.15)
plt.fill_between(np.arange(0,6,1), predictions_2011-predictions_2011_std,
                  predictions_2011+predictions_2011_std, color='#670099',alpha=0.15)
plt.ylabel('sea ice thickness [m]')
plt.xticks(np.arange(6), ['November', 'December', 'January','February','March','April'], rotation=20)  # Set text labels.
plt.grid(linestyle=':')
ax.text(0.03,0.90,'(g)',fontsize=12,transform=ax.transAxes)
plt.legend(bbox_to_anchor=(1.02, 1.02), frameon=False)

fig.tight_layout()
```
